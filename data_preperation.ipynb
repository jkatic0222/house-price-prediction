```python
import pandas as pd
```

```python
# the used data was scraped on different dates so the changes of that data
# were done separately so the code is made to import only one .csv file

raw_data = pd.read_csv('njuskalo_3.csv') # import of tha latest .csv file
```

```python
raw_data = raw_data.drop(raw_data.index[:400]) # due to problems with scraping (captcha) the first 400 rows were empty
```

```python
raw_data.head()
```

```python
# transforming and preparing data before making a new dataframe and .csv

# ad id
ad_id = []
for el in raw_data.ad_id:
    ad_id.append(el.split(': ')[1])


# location
county_legend = {'Zagrebačka': 1, 'Krapinsko-zagorska': 2, 'Sisačko-moslavačka': 3, 'Karlovačka': 4, 'Varaždinska': 5,
                'Koprivničko- križevačka': 6, 'Bjelovarsko-bilogorska': 7, 'Primorsko-goranska': 8, 'Ličko-senjska': 9,
                'Virovitičko-podravska': 10, 'Požeško-slavonska': 11, 'Brodsko-posavska': 12, 'Zadarska': 13,
                'Osječko-baranjska': 14, 'Šibensko-kninska': 15, 'Vukovarsko-srijemska': 16, 'Splitsko-dalmatinska': 17,
                'Istarska': 18, 'Dubrovačko-neretvanska': 19, 'Međimurska': 20, 'Grad Zagreb': 21}
county_number, county, location1, location2 = [], [], [], []
for el in raw_data.location:
    loc = el.split(', ')
    try: county.append(loc[0])
    except: county.append(None)
    try: county_number.append(county_legend[loc[0]])
    except: county_number.append(None)
    try: location1.append(loc[1])
    except: location1.append(None)
    try: location2.append(loc[2])
    except: location2.append(None)


# sqft
sqft, sqft_net = [], []
for el in raw_data.sqft:
    try: sqft.append(float(el.split(' ')[0].replace(',', '.')))
    except: sqft.append(None)
for el in raw_data.sqft_net:
    try: sqft_net.append(float(el.split(' ')[0].replace(',', '.')))
    except: sqft_net.append(None)


#number of rooms
number_of_rooms = []
for el in raw_data.number_of_rooms:
    if el == 'Garsonijera': number_of_rooms.append(0.0)
    elif el == '1-sobni': number_of_rooms.append(1.0)
    elif el == '2-sobni': number_of_rooms.append(2.0)
    elif el == '3-sobni': number_of_rooms.append(3.0)
    elif el == '4-sobni': number_of_rooms.append(4.0)
    elif el == '5+ sobni': number_of_rooms.append(5.0)

# apartment type
apartment_type = []
for el in raw_data.apartment_type:
    if el == 'U kući': apartment_type.append(0)
    elif el == 'U stambenoj zgradi': apartment_type.append(1)

# price
price = []
for el in raw_data.price:
    price.append(el[:-2])

for i in range(len(price)):
    price[i] = price[i].replace('.', '')
    price[i] = float(price[i].replace(',', '.'))

# number of floors
number_of_floors = []
for el in raw_data.number_of_floors:
    if el == 'Jednoetažni': number_of_floors.append(1.0)
    elif el == 'Dvoetažni': number_of_floors.append(2.0)
    elif el == 'Višeetažni': number_of_floors.append(3.0)

# date
date = []
for el in raw_data.post_date:
    date.append(pd.to_datetime(el.split(' ')[0], dayfirst=True))

data_reorg = {'ad_id': ad_id,
             'link': raw_data.link.tolist(),
             'county_number': county_number,
             'county': county,
             'location1': location1,
             'location2': location2,
             'apartment_type': apartment_type,
             'sqft(m2)': sqft,
             'sqft_net(m2)': sqft_net,
             'number_of_rooms': number_of_rooms,
             'balcony_type': raw_data.balcony_type.tolist(),
             'number_of_floors': number_of_floors,
             'floor_number': raw_data.floor_number.tolist(),
             'floors_in_building': raw_data['floors_in _building'].tolist(),
             'year_of_construction': raw_data.year_of_construction.tolist(),
             'year_of_renovation': raw_data.year_of_renovation.tolist(),
             'parking_spots': raw_data.parking_spots.tolist(),
             'energy_class': raw_data.energy_class.tolist(),
             'price(€)': price,
             'post_date': date}

df = pd.DataFrame(data_reorg)
```

```python
# saving data to a new .csv file
df.to_csv('data_njuskalo_3.csv')
```

Combining all the separate .csv files to a single one
